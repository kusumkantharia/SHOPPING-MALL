```python
def encode_and_bind(original_dataframe, feature_to_encode, column_prefix):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]] , prefix=column_prefix)
    res = pd.concat([original_dataframe, dummies], axis=1)
    return res
```

# Fitting a Decision tree model

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix, roc_curve, auc,recall_score, precision_score
```
